```python
#Declustering Weights for Porosity 

plt.subplot(221)
GSLIB.locmap_st(df,'X','Y','Por-Wts',xmin,xmax,ymin,ymax,0.0,4,'Declustering Weights','X (m)','Y (m)','Weights',cmap)

plt.subplot(222)
GSLIB.hist_st(df['Por-Wts'],0.5,4,log=False,cumul=False,bins=20,weights=None,xlabel="Weights",title="Declustering Weights")
plt.ylim(0.0,75)

plt.subplot(223)
GSLIB.hist_st(df['Porosity'],0.05,0.18,log=False,cumul=False,bins=20,weights=None,xlabel="Porosity",title="Naive Porosity")
plt.ylim(0.0,75)

plt.subplot(224)
GSLIB.hist_st(df['Porosity'],0.05,0.18,log=False,cumul=False,bins=20,weights=df['Por-Wts'],xlabel="Porosity",title="Declustered Porosity")
plt.ylim(0.0,75)

por_mean = np.average(df['Porosity'].values)
por_dmean = np.average(df['Porosity'].values,weights=df['Por-Wts'].values)

print('Porosity naive mean is ' + str(round(por_mean,3))+'.')
print('Porosity declustered mean is ' + str(round(por_dmean,3))+'.')
cor = (por_mean-por_dmean)/por_mean
print('Correction of ' + str(round(cor,4)) +'.')

print('\nSummary statistics of the declsutering weights:')
print(stats.describe(wts))

plt.subplots_adjust(left=0.0, bottom=0.0, right=2.0, top=2.5, wspace=0.2, hspace=0.2)
plt.show()
```

We will start to use bootstrap to get the uncertainty models for porosity, permeability, facies proportion and oil in place.

```python
## This is for facies 1
L = 1000                                 # set the number of realizations
mean = np.zeros(L); stdev = np.zeros(L); P25=np.zeros(L); P50=np.zeros(L); P75=np.zeros(L);   # declare arrays to hold the realizations of the statistics
for l in range(0, L):                      # loop over realizations
    samples = random.choices(df_1['Porosity'].values, weights=df_1['Por-Wts'].values, cum_weights=None, k=len(df_1))
    mean[l] = np.average(samples)
    stdev[l] = np.std(samples)
    P25[l]=np.percentile(samples,0.25)
    P50[l]=np.percentile(samples,0.5)
    P75[l]=np.percentile(samples,0.75)
plt.subplot(121)
GSLIB.hist_st(mean,0.1,0.13,False,False,50,None,'Average Porosity (fraction)','Bootstrap Uncertainty in Porosity Average')

plt.subplot(122)
GSLIB.hist_st(stdev,0.01,0.03,False,False,50,None,'Standard Deviation Porosity (fraction)','Bootstrap Uncertainty in Porosity Standard Deviation')

plt.subplots_adjust(left=0.0, bottom=0.0, right=3.0, top=1.2, wspace=0.2, hspace=0.2)
plt.show()   
    
print('Summary Statistics for Bootstrap Porosity Mean Realizations:')
print(stats.describe(mean))
print('P10 ' + str(round(np.percentile(mean,10),3)) + ', P50 ' + str(round(np.percentile(mean,50),3)) + ', P90 ' + str(round(np.percentile(mean,90),3))) 

print('\nSummary Statistics for Bootstrap Porosity Standard Deviation Realizations:')
print(stats.describe(stdev))
print('P10 ' + str(round(np.percentile(stdev,10),3)) + ', P50 ' + str(round(np.percentile(stdev,50),3)) + ', P90 ' + str(round(np.percentile(stdev,90),3))) 
```
