#### Summary Statistics for Tabular Data

The table includes X and Y coordinates (meters), Facies 1 and 2 (1 is sandstone and 0 interbedded sand and mudstone), Porosity (fraction), permeability as Perm (mDarcy) and acoustic impedance as AI (kg/m2s*10^6). 

There are a lot of efficient methods to calculate summary statistics from tabular data in DataFrames. The describe command provides count, mean, minimum, maximum, and quartiles all in a nice data table. We use transpose just to flip the table so that features are on the rows and the statistics are on the columns.

```python
df.describe().transpose()
```

#### Plotting Distributions

Let's display some histograms. I reimplimented the hist function from GSLIB. See the parameters.

```python
df_1 = df.loc[(df['Facies'] == 1)] 
#df_1.describe
```

```python
GSLIB.hist(df_1['Porosity'].values,pormin,pormax,log=False,cumul = False,bins=20,weights = None, xlabel='Porosity (fraction)',title='Porosity Well Data - Facies 1 ',fig_name='hist_Porosity')
GSLIB.hist(df_0['Porosity'].values,pormin,pormax,log=False,cumul = False,bins=20,weights = None, xlabel='Porosity (fraction)',title='Porosity Well Data -Facies 0',fig_name='hist_Porosity')
```

```python
plt.subplot(131)
GSLIB.hist_st(df['Porosity'].values,pormin,pormax,log=False,cumul = False,bins=20,weights = None,xlabel='Porosity (fraction)',title='Porosity Well Data')

plt.subplot(132)
GSLIB.hist_st(df['Perm'].values,permin,permax,log=False,cumul = False,bins=np.logspace(np.log10(0.01),np.log10(100000), 20),weights = None,xlabel='Permeability (mD)',title='Permeability Well Data')
plt.gca().set_xscale("log")
plt.subplot(133)
GSLIB.hist_st(df['AI'].values,AImin,AImax,log=False,cumul = False,bins=20,weights = None,xlabel= r'Acoustic Impedence ($kg/m^2s\times10^6$)',title='AI Well Data')

plt.subplots_adjust(left=0.0, bottom=0.0, right=3.0, top=1.5, wspace=0.1, hspace=0.2)
plt.savefig('hist_Porosity_Multiple_bins.png',dpi=600,bbox_inches="tight")
plt.show()
```

```python
# Open file of seismic AI 

s_raw = open("6_AI.csv","r")
s = []

# Reads file as 2D list.  For loops clean the text and turns them into a 1D list "s"
for i in s_raw.readlines():
	x = i.strip("\n").split(",")
	for k in x:
		s.append(float(k))

# Convert the list into an array
s_array = np.array(s)






```

```python
#histogram of AI from well data and seismic


plt.subplot(212)
plt.hist(df['AI'].values,density=True, cumulative=False, label='well',
           histtype='bar', alpha=0.5, bins = 20, color='blue', edgecolor = 'black')
plt.hist(s_array,density=True, cumulative=False, label='seismic',
            histtype='bar', alpha=.5, bins = 20, color='red', edgecolor = 'black')
plt.xlabel(r'Acoustic Impedence($kg/m^2s\times10^6$)')
plt.title('AI Frequency Distribution')
plt.ylabel('Frequency')
plt.legend(loc='upper left')
plt.subplots_adjust(left=0.0, bottom=0.0, right=1.0, top=3.0, wspace=0.1, hspace=0.2)
plt.savefig('cdf_PorPermAI.png',dpi=600,bbox_inches="tight")
plt.show()

```

```python
# CDF of AI data from well and seismic '
plt.hist(df['AI'].values,density=True, cumulative=True, label='well',
           histtype='stepfilled', alpha=0.5, bins = 5000, color='blue',  range=[1000,6000])
plt.hist(s_array,density=True, cumulative=True, label='seismic',
            histtype='stepfilled', alpha=0.5, bins = 5000, color='red', range=[1000, 6000])
plt.xlabel(r'Acoustic Impedence($kg/m^2s\times10^6$)')
plt.title('AI CDF')
plt.ylabel('Cumulative Probability')
plt.legend(loc='upper left')
plt.subplots_adjust(left=0.0, bottom=0.0, right=1.0, top=1.0, wspace=0.1, hspace=0.2)
plt.savefig('cdf_PorPermAI.png',dpi=600,bbox_inches="tight")
plt.show()
```
